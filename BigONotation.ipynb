# O(logN) 

```python
def binary_search(data, value):
    global count
    n = len(data)
    left = 0
    right = n - 1
    while left <= right:
        count=count+1
        middle = (left + right) // 2
        if value < data[middle]:
            right = middle - 1
        elif value > data[middle]:
            left = middle + 1
        else:
            return middle
        
```

```python
x=[]
y=[]
for i in range(1,10000):
    count=0
    array=list(range(0,i))
    result=binary_search(array,array[len(array)-1])
    x.append(i)
    y.append(count)
    
plt.plot(x,y);
plt.xlabel("Input boyutu");
plt.ylabel("Gerceklestirilen islem sayisi");
```
